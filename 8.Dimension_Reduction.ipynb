```python
# wine data 불러오기 
import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

dat_wine=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/'
                     'wine/wine.data',header=None)
dat_wine.head()
dat_wine.columns = ['class label', 'alchohol', 'malic acid', 'ash', 
                    'alcalinity of ash', 'magnesium', 'total phenols', 
                    'flavanoids', 'nonflavanoid phenols', 
                    'proanthocyanins', 'color intensity', 'hue', 
                    'OD208', 'proline']  # Column names
print('class label:', np.unique(dat_wine['class label']))  # Class 출력
y = dat_wine['class label']
X = dat_wine.drop("class label", axis = 1)

std=StandardScaler()
X_std = std.fit_transform(X)

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=1,stratify=y)

X_train_std=std.fit_transform(X_train)
X_test_std=std.transform(X_test)
```

```python
# MDS에 의한 차원 축소
from sklearn.manifold import MDS
mds = MDS(n_components=3, random_state=2)
X_mds = mds.fit_transform(X_std)

import plotly.graph_objects as go
fig=go.Figure() 
fig.add_trace(go.Scatter3d(x=X_mds[:,0],y=X_mds[:,1], z = X_mds[:,2],opacity=0.7,mode='markers',
                                           marker=dict(size=3, color=y,colorscale='Viridis'))) 
fig.show()

```

```python
mds = MDS(n_components=3, random_state=2)
X_mds=mds.fit_transform(X_std)

X_train_mds, X_test_mds, y_train, y_test = train_test_split(X_mds, y, test_size=0.3,
                                                    random_state=1, stratify=y)

lr = LogisticRegression(random_state =1, multi_class= 'ovr')
lr.fit(X_train_mds, y_train)
y_train_pred_mds=lr.predict(X_train_mds)
y_test_pred_mds=lr.predict(X_test_mds)

# Logistic 적합
print('For MDS:')
print('LogReg train accuracy:',np.around(
    accuracy_score(y_train, y_train_pred_mds), 2))
print('LogReg test accuracy:',np.around(
    accuracy_score(y_test, y_test_pred_mds), 2))

cld = LinearDiscriminantAnalysis(store_covariance = True)
cld.fit(X_train_mds, y_train)
y_train_pred_mds1=cld.predict(X_train_mds)
y_test_pred_mds1=cld.predict(X_test_mds)

# LDA 적합
print('LDA train accuracy:',np.around(
    accuracy_score(y_train, y_train_pred_mds1), 2))
print('LDA test accuracy:',np.around(
    accuracy_score(y_test, y_test_pred_mds1), 2))
```

```python
# ISOMAP에 의한 차원 축소
from sklearn.manifold import Isomap
iso=Isomap(n_neighbors=50, n_components=3)
iso_out=iso.fit_transform(X_std)
fig=go.Figure() 
fig.add_trace(go.Scatter3d(x=iso_out[:,0],y=iso_out[:,1],z=iso_out[:,2],mode='markers',
                                           marker=dict(size=3, color=y,colorscale='Viridis'))) 
fig.show()
```

```python
perplexity_value = list(np.arange(2, 50,1))
train_accuracy_LDA = []
test_accuracy_LDA = []
for i in perplexity_value :
    tsne = TSNE(n_components = 3, perplexity = i)
    x_tsne = tsne.fit_transform(X_std)
    X_train, X_test, y_train, y_test = train_test_split(x_tsne, y, test_size=0.3,
                                                    random_state=1, stratify=y)
    cld = LinearDiscriminantAnalysis(store_covariance = True)
    cld.fit(X_train, y_train)
    y_test_pred = cld.predict(X_test)
    y_train_pred = cld.predict(X_train)
    test_accuracy_LDA.append(accuracy_score(y_test, y_test_pred))
    train_accuracy_LDA.append(accuracy_score(y_train, y_train_pred))
```

```python
test_accuracy_lr = []
train_accuracy_lr = []
for i in perplexity_value :
    tsne = TSNE(n_components = 3, perplexity = i)
    x_tsne = tsne.fit_transform(X_std)
    X_train, X_test, y_train, y_test = train_test_split(x_tsne, y, test_size=0.3,
                                                    random_state=1, stratify=y)
    lr = LogisticRegression(random_state =1, multi_class= 'ovr')
    lr.fit(X_train, y_train)
    y_train_pred = lr.predict(X_train)
    y_test_pred = lr.predict(X_test)
    test_accuracy_lr.append(accuracy_score(y_test, y_test_pred))
    train_accuracy_lr.append(accuracy_score(y_train, y_train_pred))
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=perplexity_value, y = train_accuracy_LDA, mode = 'lines', name = 'Acc. Train'))
fig.add_trace(go.Scatter(x=perplexity_value, y = test_accuracy_LDA, mode = 'lines', name = 'Acc. Test'))

fig.update_layout(title = 'LDA cls over Various Perplexity Values',yaxis_title = 'Accuracy')
fig.show()
```
