```
# 데이터 불러오기
import seaborn as sns # seaborn을 불러오고 SNS로 축약
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder    # LabelEncoder() method를 불러옴
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

iris = sns.load_dataset('iris')  # iris라는 변수명으로 Iris data를 download
X = iris.drop('species', axis=1) # 'species'열을 drop하고 input X를 정의
y = iris['species']
print(np.unique(y))

# y data를 범주형으로 변환
classle = LabelEncoder() # one-hot과 구분해서 사용
y = classle.fit_transform(iris['species'].values) # species 열의 문자열을 categorical 값으로 전환

# 전체 data를 training set과 test set으로 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    random_state=1, stratify=y) #층화추출법
print(X_train.shape, X_test.shape)
```

```
# Logistic regression (one-versus-rest)
Logit = LogisticRegression(C=1e2, random_state=1, multi_class = 'ovr')
Logit.fit(X_train, y_train)
y_train_pred = Logit.predict(X_train)
y_test_pred = Logit.predict(X_test)

# Accuracy score
print(accuracy_score(y_test,y_test_pred))    # test accuracy
print(accuracy_score(y_train,y_train_pred))  # train accuracy
print(confusion_matrix(y_test, y_test_pred)) # Confusion matrix
```

```
y_test_pred_proba = Logit.predict_proba(X_test)
print('Predicted Classes:', np.around(y_test_pred[:5],3))
print('Predicted Probabilities:', np.around(y_test_pred_proba[0],3))
np.sum(y_test_pred_proba[:5], axis = 1)
```

```
def softmax(x):
    e_x = np.exp(x)
    return e_x / e_x.sum()

def sigmoid(x):
    e_x = np.exp(-x)
    return 1 / (1+e_x)
```
