```
import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold

dat_wine=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/'
                     'wine/wine.data',header=None)
dat_wine.head()
dat_wine.columns = ['class label', 'alchohol', 'malic acid', 'ash', 
                    'alcalinity of ash', 'magnesium', 'total phenols', 
                    'flavanoids', 'nonflavanoid phenols', 
                    'proanthocyanins', 'color intensity', 'hue', 
                    'OD208', 'proline']  # Column names
print('class label:', np.unique(dat_wine['class label']))  # Class 출력
y = dat_wine['class label']
X = dat_wine.drop("class label", axis = 1)                

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,
                                                    stratify=y, random_state=1)
```

```

inner_cv=KFold(n_splits=3, shuffle=True, random_state=0)
outer_cv=KFold(n_splits=5, shuffle=True, random_state=0)

# SVM
pipe_svc = make_pipeline(StandardScaler(), SVC(random_state=1))
param_range = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]
param_grid = [{'svc__C': param_range, 'svc__kernel': ['linear']},
              {'svc__C': param_range, 'svc__gamma': param_range,
               'svc__kernel': ['rbf']}]
gs = GridSearchCV(estimator=pipe_svc, param_grid=param_grid,
                  scoring='accuracy', cv=inner_cv)

#CV Score 추출 및 최적 hyperparameter 찾기
scores = cross_val_score(gs, X, y, scoring='accuracy', cv=outer_cv)
print('For SVM:')
print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
print('Each accuracy:', np.round(scores, 3))
gs = gs.fit(X_train, y_train)
print('Best accuracy: %.3f' %gs.best_score_)
print('Best parameter:' ,gs.best_params_)

# BEST 인걸 뽑아서 train 전체 데이터에 적용
clf = gs.best_estimator_
clf.fit(X_train, y_train)
print('Train acc = %.3f' %clf.score(X_train,y_train))
print('Test acc = %.3f' %clf.score(X_test, y_test))
```

```

inner_cv=KFold(n_splits=3, shuffle=True, random_state=0)
outer_cv=KFold(n_splits=5, shuffle=True, random_state=0)

# Decision tree
pipe_cart = make_pipeline(StandardScaler(), DecisionTreeClassifier(random_state=1))
gs = GridSearchCV(estimator=pipe_cart, param_grid=[{'decisiontreeclassifier__max_depth': 
                    [1, 2, 3, 4, 5, 6, 7, 8,None]}], scoring='accuracy', cv=inner_cv)

#CV Score 추출 및 최적 hyperparameter 찾기
scores = cross_val_score(gs, X, y, scoring='accuracy', cv=outer_cv)
print('For CART:')
print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
print('Each accuracy:', np.round(scores, 3))

gs = gs.fit(X_train, y_train)
print('Best accuracy: %.3f' %gs.best_score_)
print('Best parameter:' ,gs.best_params_)

# BEST 인걸 뽑아서 train 전체 데이터에 적용
clf = gs.best_estimator_
clf.fit(X_train, y_train)
print('Train acc = %.3f' %clf.score(X_train,y_train))
print('Test acc = %.3f' %clf.score(X_test, y_test))
```
